### Facebook Data -- Data Science Work


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('facebook.csv')
```

```python
df.head()
```

```python
# Age 
df.describe()
```

```python
# Drop the 'userid' column as it is not required
df.drop(columns=['userid'], inplace=True)
```

```python
# Check for missing values
df.isna().sum()
```

```python
# Count the number of people in each age group
age_counts = df['age'].value_counts()
print(age_counts)
```

```python
# Now Visualize it
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Age between  13 to 70
df[df['age'].between(13, 70)]['age'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of People', fontsize=12)
plt.title('Age Frequency Distribution', fontsize=14, fontweight='bold')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 18 Age Peoples are the most

```python
# friend_count by age group by 
# tell which age has highest average friend_count
# We wil Consider Records that are from 13 to 70 age only

df[df['age'].between(13, 70)].groupby('age')['friend_count'].mean()
```

```python
# Now Visualize it
age_friend=df[df['age'].between(13, 70)].groupby('age')['friend_count'].mean()

plt.figure(figsize=(12, 6))
age_friend.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Average Friend Count', fontsize=12)
plt.title('Average Friend Count by Age', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(age_friend.idxmax())
```

```python
# From this 16 year old how many are male and female
df[df['age'] == 16]['gender'].value_counts()

# PIE CHART
gender_counts = df[df['age'] == 16]['gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Gender Distribution for 16 Year Olds')
plt.show()
```

### 16 Age People has the most friends

```python
# likes by age group by 
# tell which age has highest average likes
# We wil Consider Records that are from 13 to 70 age only

df[df['age'].between(13, 70)].groupby('age')['likes'].mean()
```

```python
# Now Visualize it
age_likes=df[df['age'].between(13, 70)].groupby('age')['likes'].mean()

plt.figure(figsize=(12, 6))
age_likes.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Average Likes', fontsize=12)
plt.title('Average Likes by Age', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(age_likes.idxmax())
```

```python
# From this 17 year old how many are male and female
df[df['age'] == 17]['gender'].value_counts()

# PIE CHART
gender_counts = df[df['age'] == 17]['gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Gender Distribution for 17 Year Olds')
plt.show()
```
