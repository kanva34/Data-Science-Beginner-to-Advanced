```python
import numpy as np
from scipy import stats
```

```python
# Example: Observed vs Expected frequencies
# Suppose a dice is rolled 6 times and we record observed counts
observed = np.array([22, 17, 20, 26, 22, 13])
expected = np.array([20, 20, 20, 20, 20, 20])  # Expected frequency if fair dice
```

```python
# Step 1: Calculate Chi-Square statistic
chi_square_stat = np.sum((observed - expected)**2 / expected)
print("Calculated Chi-Square value:", round(chi_square_stat, 3))
```

```python
# Step 2: Degrees of freedom (number of categories - 1)
df = len(observed) - 1
print("Degrees of freedom:", df)
```

```python
# Step 3: Critical Chi-Square value at alpha = 0.05
alpha = 0.05
chi_square_critical = stats.chi2.ppf(1 - alpha, df)
print("Critical Chi-Square value (alpha=0.05):", round(chi_square_critical, 3))
```

```python
# Step 4: Decision
if chi_square_stat > chi_square_critical:
    print("Reject the null hypothesis (H0): The dice may not be fair")
else:
    print("Fail to reject the null hypothesis (H0): No evidence against fairness")
```
