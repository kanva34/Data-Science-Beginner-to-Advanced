```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
dataset = pd.read_csv("titanic.csv")
```

```python
# Display the first few rows of the dataset
dataset.head(3)
```

```python
# Fill missing values in 'Age' column with the median age
dataset["Age"].fillna(dataset["Age"].median(), inplace=True)
```

```python
# Display the 'Age' column after filling missing values
dataset["Age"]
```

```python
# Show descriptive statistics for all numeric columns
dataset.describe()
```

```python
# Create a boxplot to visualize the distribution of 'Age' and detect outliers
plt.figure(figsize=(8,5))
sns.boxplot(x="Age", data=dataset, color="lightgreen")
plt.title("Boxplot of Passenger Ages")
plt.xlabel("Age")
plt.show()
```

```python
# Calculate the 25th percentile (Q1) and 75th percentile (Q3) of 'Age'
q1 = np.percentile(dataset["Age"], 25)
q3 = np.percentile(dataset["Age"], 75)
print(f"25th Percentile (Q1): {q1}")
print(f"75th Percentile (Q3): {q3}")
```
