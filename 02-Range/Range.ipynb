# This notebook demonstrates how to calculate the **range** 


```python
# --- Step 1: Import Libraries ---
import pandas as pd
```

```python
# --- Step 2: Load Dataset ---
dataset = pd.read_csv("tips.csv")
```

```python
# Preview first 3 rows
dataset.head(3)
```

```python
# --- Step 3: Calculate Minimum and Maximum ---
min_total_bill = dataset["total_bill"].min()
max_total_bill = dataset["total_bill"].max()
```

```python
print(f"Minimum total bill: {min_total_bill}")
print(f"Maximum total bill: {max_total_bill}")
```

```python
# --- Step 4: Calculate Range ---
total_bill_range = max_total_bill - min_total_bill
print(f"Range of total bill: {total_bill_range}")
```
