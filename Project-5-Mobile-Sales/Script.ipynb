## Mobile Phone Sales Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Sales.csv')
```

```python
df.head()
```

```python
df.size
```

```python
# Checking Which Brands Data is This
df['Brands'].unique()
```

```python
# Brand Wise Check how many Records are then Pieploting it
brand_counts = df['Brands'].value_counts()
print(brand_counts)
plt.figure(figsize=(8, 8))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Brand Distribution')
plt.show()
```

```python
# Average Price of Each Brand Mobile Saled in this Dataset
average_prices = df.groupby('Brands')['Selling Price'].mean()
print(average_prices)
plt.figure(figsize=(10, 6))
average_prices.plot(kind='bar', color='skyblue')
plt.title('Average Price of Each Brand')
plt.xlabel('Brands')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Max Discounts Give on which Brand Mobiles and then plot them 
max_discounts = df.groupby('Brands')['discount percentage'].max().sort_values(ascending=False)
print(max_discounts)

plt.figure(figsize=(10, 6))
max_discounts.plot(kind='bar', color='orange')
plt.title('Maximum Discounts by Brand')
plt.xlabel('Brands')
plt.ylabel('Maximum Discount Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

```python
# Which Mobiles Brands have the highest rating
highest_rated = df.groupby('Brands')['Rating'].max().sort_values(ascending=False)   
print(highest_rated)

plt.figure(figsize=(10, 6))
highest_rated.plot(kind='bar', color='green')
plt.title('Highest Rated Mobile Brands')
plt.xlabel('Brands')
plt.ylabel('Highest Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Highest Rated Apple Mobiles
apple_highest_rated = df[df['Brands'] == 'Apple'].sort_values(by='Rating', ascending=False).head(5)
print(apple_highest_rated)

plt.figure(figsize=(10, 6))
plt.bar(apple_highest_rated['Models'], apple_highest_rated['Rating'],
        color='purple')
plt.title('Top 5 Highest Rated Apple Mobiles')
plt.xlabel('Model')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Highest Rated Samsung Mobiles
samsung_highest_rated = df[df['Brands'] == 'SAMSUNG'].sort_values(by='Rating', ascending=False).head(5)
print(samsung_highest_rated)    

plt.figure(figsize=(10, 6))
plt.bar(samsung_highest_rated['Models'], samsung_highest_rated['Rating'],
        color='red')
plt.title('Top 5 Highest Rated Samsung Mobiles')
plt.xlabel('Model')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Checking any Missing Values in the Dataset
df.isnull().sum()
```

```python
# Fill Rating with Average Rating
df['Rating'].fillna(df['Rating'].mean(), inplace=True)

```

```python
df.isnull().sum()
```

```python
# Remove all the records completely whose Memory or Storage is missing
df.dropna(subset=['Memory', 'Storage'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.size
```

```python
df.head()
```

```python
brand_summary = df.groupby("Brands").agg(
    avg_price=("Selling Price", "mean"),
    avg_rating=("Rating", "mean"),
    avg_discount_pct=("discount percentage", "mean"),
    model_count=("Models", "count")
).sort_values("avg_price", ascending=False)

print(brand_summary)
```

```python
price_rating_corr = df["Selling Price"].corr(df["Rating"])
print("PriceRating Correlation:", round(price_rating_corr, 2))

```

```python
discount_rating_corr = df["discount percentage"].corr(df["Rating"])
discount_rating_corr
```

```python
# Mid Level Mobiles having GOod Rating and Average Price as compare to others
```

```python
df["Value_Score"] = df["Rating"] / df["Selling Price"]
df.sort_values("Value_Score", ascending=False)[
    ["Mobile","Selling Price","Rating","Value_Score"]
].head(15)
```

```python
# 3 Discount vs Rating Correlation
corr = df["Selling Price"].corr(df["Rating"])
print(f"Correlation between Selling Price and Rating: {corr:.2f}")

# Scatter plot Selling Price vs Rating
plt.figure(figsize=(10,4))
sns.scatterplot(data=df, x="Selling Price", y="Rating", hue="Brands")
plt.title("Selling Price vs Rating")
plt.show()
```
