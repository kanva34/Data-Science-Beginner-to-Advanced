# One-Sample T-Test

## Objective
The objective of this test is to determine whether the sample mean is significantly less than the hypothesized population mean using a one-sample T-test.

```python
import numpy as np
from scipy import stats
```

```python
# Given values
sample_mean = 148        # Mean of the sample (x̄)
population_mean = 150   # Hypothesized population mean (μ)
sample_std = 5          # Sample standard deviation (s)
n = 25                  # Sample size
alpha = 0.05            # Significance level
```

```python
# Degrees of freedom
df = n - 1
```

```python
# Step 1: Calculate the t statistic
t_calculated = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
print("Calculated t value:", round(t_calculated, 3))
```

```python
# Step 2: Find the critical t value (left-tailed test)
t_critical = stats.t.ppf(alpha, df)
print("Critical t value:", round(t_critical, 3))
```

```python
# Step 3: Decision rule
if t_calculated < t_critical:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```
