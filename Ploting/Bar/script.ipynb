### Bar Plot
A Bar Plot (or Bar Chart) shows categorical data using rectangular bars.
Each barâ€™s height represents a value (like average, count, or sum).

Use it when you want to:

1. Compare different categories (e.g., cities, house types)

2. Show counts or averages

3. See which category dominates

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
values = [10, 23, 17, 5]

plt.bar(categories, values)
plt.title("Simple Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
import pandas as pd

shoppings = pd.read_csv("Shoppings.csv")
shoppings.head()
```

```python
maleCount = shoppings[shoppings["Gender"] == "Male"].shape[0]
maleCount
```

```python
femaleCount=shoppings[shoppings["Gender"]=="Female"].shape[0]
femaleCount
```

```python
# Bar Plot
plt.bar(["Male", "Female"], [maleCount, femaleCount], color=['blue', 'pink'])
plt.title("Number of Shoppers by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Shoppers")
plt.show()
```

```python
# Bar Plot by Category wise

categories = shoppings['Category'].unique()
categories  
```

```python
categories_counts = shoppings['Category'].value_counts()
categories_counts
```

```python
plt.bar(categories_counts.index, categories_counts.values, color='blue')
plt.title("Number of Shoppers by Category")
plt.xlabel("Category")
plt.ylabel("Number of Shoppers")
plt.show()
```

```python
categories_counts = shoppings['Item Purchased'].value_counts()
categories_counts
```

```python

plt.figure(figsize=(8, 5))   # ðŸ’¡ width=12, height=6 (you can adjust)
plt.bar(categories_counts.index, categories_counts.values, color='green')

plt.title("Number of Shoppers by Category")
plt.xlabel("Category")
plt.ylabel("Number of Shoppers")
plt.xticks(rotation=45)  # tilt labels if long
plt.tight_layout()       # adjust spacing
plt.show()
```
