# # ðŸ“Š Section Scores Analysis
# This notebook calculates **Mean, Variance, Standard Deviation (SD)** 
# and **Mean Absolute Deviation (MAD)** for two sections of student scores.

```python
# --- Step 1: Import Libraries ---
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Step 2: Define Scores ---
sec_a = np.array([75, 65, 73, 68, 72, 76])
sec_b = np.array([90, 47, 43, 96, 93, 51])
student_no = np.array([1, 2, 3, 4, 5, 6])
```

```python
# --- Step 3: Calculate Standard Deviation ---
std_a = np.std(sec_a)
std_b = np.std(sec_b)
print(f"Standard Deviation of Sec A: {std_a:.2f}")
print(f"Standard Deviation of Sec B: {std_b:.2f}")
```

```python
# --- Step 4: Calculate Variance ---
var_a = np.var(sec_a)
var_b = np.var(sec_b)
print(f"Variance of Sec A: {var_a:.2f}")
print(f"Variance of Sec B: {var_b:.2f}")
```

```python
# --- Step 5: Calculate Mean ---
mean_b = np.mean(sec_b)  # using Sec B as reference
print(f"Mean of Sec B: {mean_b:.2f}")
```

```python
# --- Step 6: Calculate Mean Absolute Deviation (MAD) ---
mad_a = np.mean(np.abs(sec_a - mean_b))
mad_b = np.mean(np.abs(sec_b - mean_b))
print(f"MAD of Sec A (from mean of Sec B): {mad_a:.2f}")
print(f"MAD of Sec B (from mean of Sec B): {mad_b:.2f}")
```

```python
# --- Step 7: Visualization ---
plt.figure(figsize=(10, 3))
plt.scatter(sec_a, student_no, label="Sec A", s=80)
plt.scatter(sec_b, student_no, color="red", label="Sec B", s=80)
plt.plot([mean_b]*len(student_no), student_no, c="blue", label="Mean of Sec B", linewidth=2)
plt.xlabel("Scores")
plt.ylabel("Student No")
plt.title("Section Scores with Mean Line")
plt.legend()
plt.show()
```
