```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Generate Random data for population
pop_data= [np.random.randint(10,100) for _ in range(10000)]
pop_table= pd.DataFrame({"pop_data":pop_data})

```

```python
# Check mean
np.mean(pop_data)
```

```python
# Check distribution of data
plt.figure(figsize=(3,2))
sns.kdeplot(x="pop_data",data=pop_table)
plt.show()
```

```python
# 10 % data
# (10000*10)/100
```

```python
# Generate 60 samples of size 500
sample_mean = []
for num_sample in range(60):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data,))
    sample_mean.append(np.mean(sample_data))
```

```python
# Check mean of samples
np.mean(sample_mean)
```

```python
sample_M = pd.DataFrame({"Sample_Mean":sample_mean})
```

```python
# Check distribution
plt.figure(figsize=(3,2))
sns.kdeplot(x="Sample_Mean",data=sample_M)
plt.show()
```
