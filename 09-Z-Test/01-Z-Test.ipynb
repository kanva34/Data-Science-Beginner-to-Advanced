# One-Sample Z-Test

## Objective
The objective of this test is to determine whether the sample mean is significantly greater than the population mean using a one-sample Z-test.

```python
# Import necessary libraries
import numpy as np
from scipy import stats
```

```python
# Given values
sample_mean = 90        # Mean of the sample (x̄)
population_mean = 82    # Mean of the population (μ)
population_std = 20     # Population standard deviation (σ)
n = 81                  # Sample size
alpha = 0.05            # Significance level
```

```python
# Step 1: Calculate the Z statistic
z_calculated = (sample_mean - population_mean) / (population_std / np.sqrt(n))
print("Calculated Z value:", round(z_calculated, 3))
```

```python
# Step 2: Find the critical Z value (right-tailed test)
z_critical = stats.norm.ppf(1 - alpha)
print("Critical Z value:", round(z_critical, 3))
```

```python
# Step 3: Decision rule
if z_calculated > z_critical:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```
