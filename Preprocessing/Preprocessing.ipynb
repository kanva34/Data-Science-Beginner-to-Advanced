#### Installing All Dependencies

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

### Check any Null Values

```python
df.isna().sum()
```

### Already Records are less we wil not drop the Complete Row , We will Fill it with a values

```python
df['Age'].describe()
```

```python
## Let Worst Fill Missing Values of Age
df['Age'].fillna(df['Age'].mean(),inplace=True)

## Why we Choosed Mean?
## Because Age is Normally Distributed
```

### As the Whole Titanic is Drown , The Cabin does not tell anything meaninfull so we will fill it with 
### Known & Unknown

```python
## If Value exists then known and otherwise unknown
df['Cabin']=df['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
## Again Check Missing Values
df.isna().sum() 
```

## Filling Embarked with Median more repeated place

```python
## Let Worst Fill Missing Values of Embarked
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)

## Why we Choosed Mode?
## Because Embarked is Categorical Distributed
```

```python
## Again Check Missing Values
df.isna().sum() 
```

### Good All Missing Values Removed , Now Next Step is to Make the Categorical Values to Numerical 

## Converting Text to Numbers

```python
# We will use Lablel Encoding
from sklearn.preprocessing import LabelEncoder
sexLe=LabelEncoder()
embarkedLe=LabelEncoder()


df['Sex']=sexLe.fit_transform(df['Sex'])
df['Embarked']=embarkedLe.fit_transform(df['Embarked'])
```

```python
df.head()
```

## Selecting Important Features

```python
df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
df.head()
```

## Feature Scaling Because Age is Fare is Very High as Comapre to Other Columns 


```python
# Feature Scaling will not change the Distribution of Data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])

```

```python
df.head()
```

```python
# Sample Inverse Transform

scaler.inverse_transform(df[['Age','Fare']].head())
```

### Exporting to CSV

```python
df.to_csv('Preprocessed_Titanic_Dataset.csv',index=False)
```
