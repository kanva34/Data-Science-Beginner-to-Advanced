```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
dataset = pd.read_csv("titanic.csv")
dataset.head(3)
```


# Calculate Mean, Median, Mode



```python
# Mean of Fare
mean_fare = dataset["Fare"].mean()
```

```python
# Median of Fare
median_fare = dataset["Fare"].median()
```

```python
# Mode of Fare (first value because mode can return multiple values)
mode_fare = dataset["Fare"].mode()[0]
```

```python
# Print values
print("Mean of Fare:", mean_fare)
print("Median of Fare:", median_fare)
print("Mode of Fare:", mode_fare)
```


# Visualization


```python
# Create histogram of Fare
sns.histplot(
    x="Fare",
    data=dataset,
    bins=[i for i in range(0, 81, 10)]
)

# Draw vertical lines for Mean, Median and Mode
plt.plot([mean_fare]*300, range(300), color="green", label="Mean")
plt.plot([median_fare]*300, range(300), color="blue", label="Median")
plt.plot([mode_fare]*300, range(300), color="red", label="Mode")
# Add legend
plt.legend()
# Show plot
plt.show()
```
