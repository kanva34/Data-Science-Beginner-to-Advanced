```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
dataset = pd.read_csv("titanic.csv")
```

```python
# Display the first 3 rows to inspect the data
dataset.head(3)
```

```python
# Calculate the variance of the 'Age' column
age_variance = dataset["Age"].var()
print(f"Age Variance: {age_variance}")
```

```python
# Calculate the standard deviation of the 'Age' column
age_std = dataset["Age"].std()
print(f"Age Standard Deviation: {age_std}")
```

```python
# Display descriptive statistics for the 'Age' column
dataset["Age"].describe()
```

```python
# Plot a histogram to visualize the distribution of passenger ages
plt.figure(figsize=(10,6))
sns.histplot(
    data=dataset,
    x="Age",
    bins=30,        # Number of bars in the histogram
    kde=True,       # Add a smooth density curve
    color="skyblue" # Set the color of the bars
)

# Add title and axis labels
plt.title("Distribution of Passenger Ages on Titanic")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

# Add vertical lines for mean and median
plt.axvline(dataset["Age"].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(dataset["Age"].median(), color='green', linestyle='-', label='Median')

# Add a legend to distinguish mean and median
plt.legend()

# Show the plot
plt.show()
```
